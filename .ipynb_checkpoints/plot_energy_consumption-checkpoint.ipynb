```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import dataclasses
import copy

@dataclasses.dataclass
class NodeEnergyConsumption:
    id: int
    energy: float
        
csv_file = "/Users/tozastation/workspace/ns-3.30/scratch/energy_consumption.csv"
nodes = []
node_energy_consumption = pd.read_csv(csv_file)

for index, row in node_energy_consumption.iterrows():
    node = NodeEnergyConsumption(int(row[0]), round(float(row[1]), 6))
    nodes.append(node)

fig = plt.figure(figsize=(10, 10), dpi=100)
ax = fig.add_subplot(111)
ax.set_title("node energy consumption")
ax.set_xlabel("node id")
ax.set_ylabel("energy consumption")

for node in nodes:
    ax.bar(node.id, node.energy)
    
fig.savefig("/Users/tozastation/workspace/ns-3.30/scratch/energy_consumption.png")
```
